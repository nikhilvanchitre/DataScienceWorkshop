```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans
```

```python
ds= pd.read_table("G:\Python\ml-1m\mt.dat", sep=",", names=['x','y'])
ds
```

```python
plt.scatter(ds.x,ds.y)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(ds)
plt.scatter(ds.x,ds.y)
plt.show()
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(centroids)
print(labels)
```

```python
colors=["r.","g.","b."]
for i in range(len(ds)):
    print("co-ordinate:",ds[i-1 : i],"label:",labels[i])
    plt.plot(ds.x[i],ds.y[i], colors[labels[i]], markersize = 10)

 

   
```

```python
plt.scatter(centroids[:,0],centroids[:,1], marker = "x", s=150, linewidths =5, zorder=10)
plt.show()
```
